```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
x,y = data.data, data.target
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  
df
```

```python
correlation = df.corr()['target']
correlation
```

```python
correlation.abs().sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split

x = df[['petal length (cm)','petal width (cm)', 'sepal length (cm)', 'sepal width (cm)']]
y =df[['target']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 42)

print(len(y_test))
print(len(x_test))
print(len(y_train))
print(len(x_train))
```

```python
import numpy as np
import scipy.stats as stats
data = np.array([1,2,4,57,10,20])
absolute_zscore = np.abs(stats.zscore(data))
absolute_zscore
```

```python
Threshold = 3
outlier = np.where(absolute_zscore > Threshold)[0]
outlier
```
